```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# read dataset
customers_df = pd.read_csv("customers_data.csv")
```

```python
# view dataset head
print(f"Data Head: \n{customers_df.head()}\n")
```

```python
# view dataset tail
print(f"Data Tail: \n{customers_df.tail()}\n")
```

```python
# view dataset shape
print(f"Data Shape: \n{customers_df.shape}\n")
```

```python
# view dataset info
print(f"Data Info: \n{customers_df.info()}\n")
```

```python
# view describe
print(f"Data describe: \n{customers_df.describe()}\n")
```

```python
# missing values
print(f"Missing values in each variable: \n{customers_df.isnull().sum()}\n")
```

```python
# data types
print(f"Data Type: \n{customers_df.dtypes}\n")
```

```python
# duplicates
print(f"Duplicated rows: {customers_df.duplicated().sum()}\n")
```

```python
# dataset hist
customers_df.hist()
plt.show()
```

```python
# count genre
sns.countplot(x="Genre", data=customers_df)
plt.title('Genre count')
plt.show()
```

```python
# scatter age annual income, hue genre
sns.scatterplot(x='Age', y='Annual Income (k$)', data=customers_df, hue='Genre', palette='tab10')
plt.title('Scatter Age vs Annual Income (k$)')
plt.show()
```

```python
# scatter age Spending Score, hue genre
sns.scatterplot(x='Age', y='Spending Score (1-100)', data=customers_df, hue='Genre', palette='tab10')
plt.title('Scatter Age vs Spending Score (1-100)')
plt.show()
```

```python
# scatter annual income pending Score, hue genre
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=customers_df, hue='Genre', palette='tab10')
plt.title('Scatter Annual Income (k$) vs Spending Score (1-100)')
plt.show()
```

```python
# pairplot
sns.pairplot(customers_df, x_vars=["Age", "Annual Income (k$)", "Spending Score (1-100)"],
             y_vars=["Age", "Annual Income (k$)", "Spending Score (1-100)"],
             hue="Genre",
             kind="scatter",
             palette="YlGnBu",
             height=2,
             plot_kws={"s": 35, "alpha": 0.8})
plt.title('Pairplot for customers data with genre hue')
plt.show()
```

```python
# correlation matrix
correlation_matrix = customers_df.corr().round(2)
plt.figure(figsize=(10, 10))
sns.heatmap(data=correlation_matrix, annot=True, cmap="YlGnBu")
plt.show()
```
