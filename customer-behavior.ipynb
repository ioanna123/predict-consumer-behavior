```python
customers_df1=customers_df[["Genre","Spending Score (1-100)"]]
```

```python
print(customers_df1.head())
```

```python
customers_df_scaled = customers_df1
from sklearn.preprocessing import StandardScaler

customers_df_scaled[customers_df_scaled.columns] = StandardScaler().fit_transform(customers_df_scaled)
customers_df_scaled.describe()
customers_df_scaled.head()
```

```python
customers_df_scaled.corr()
```

```python
from sklearn.cluster import KMeans
```

```python
#The Elbow Method
wcss = []
clusters = 10
for i in range(1, clusters+1):
    km = KMeans(n_clusters=i,init='k-means++', max_iter=10, n_init=10, random_state=0)
    km.fit(customers_df_scaled)
    wcss.append(km.inertia_)
```

```python
plt.figure(1 , figsize = (15 ,6)) # set dimension of image
plt.plot(range(1,clusters+1) , wcss ) # Mark the points with a solid circle
plt.xlabel('Number of Clusters') , plt.ylabel('WCSS') # label the x and y axes
```

```python
from kneed import KneeLocator    
kl = KneeLocator(
    range(1, clusters+1), wcss, curve="convex", direction="decreasing"
)

kl.elbow
```

```python
#The Silhouette Method
from sklearn.metrics import silhouette_score

sil = []

# dissimilarity would not be defined for a single cluster, thus, minimum number of clusters should be 2
for k in range(2,clusters+1):
  kmeans = KMeans(n_clusters = k, init='k-means++', max_iter=10, n_init=10, random_state=0).fit(customers_df_scaled)
  labels = kmeans.labels_
  sil.append(silhouette_score(customers_df_scaled, labels, metric = 'euclidean'))
```

```python
plt.figure(1 , figsize = (15 ,6)) # set dimension of image
plt.plot(range(2,clusters+1) , sil , 'o') # Mark the points with a solid circle
plt.plot(range(2,clusters+1) , sil , '-' , alpha = 0.5) # connect remaining points with a line
plt.xlabel('Number of Clusters') , plt.ylabel('Silhouette') # label the x and y axes
```

```python
# Kmeans algorithm
# n_clusters: Number of clusters. 
# init: k-means++. Smart initialization
# max_iter: Maximum number of iterations of the k-means algorithm for a single run
# n_init: Number of time the k-means algorithm will be run with different centroid seeds. 
# random_state: Determines random number generation for centroid initialization.
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=10, n_init=10, random_state=0)
```

```python
# Fit and predict 
y_means = kmeans.fit_predict(customers_df_scaled)
```

```python
customers_df['Label_based_Genre_SC']=y_means
```

```python
kmeans.labels_
```

```python
# Final locations of the centroid
kmeans.cluster_centers_
```

```python
 # The number of iterations required to converge
kmeans.n_iter_
```

```python
# The lowest SSE value
kmeans.inertia_
```
